```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Define data
years = ['2001', '2003', '2005', '2007', '2009']
sachin_runs = [1200, 1100, 1300, 1000, 1250]
kohli_runs = [0, 0, 600, 1050, 1200]  
sehwag_runs = [1000, 1150, 1100, 950, 900]

# Step 2: Set bar positions
x = np.arange(len(years))
width = 0.25

# Step 3: Plot side-by-side bars
plt.figure(figsize=(10, 6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

# Step 4: Labels and legend
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison: Sachin vs Kohli vs Sehwag')
plt.xticks(x, years)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

bar_categories = ['Pallavi', 'Khushi', 'Samarth', 'Yadvik']
bar_values = [10, 15, 7, 12]

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
scatter_colors = np.random.rand(50)

hist_data = np.random.randn(1000)

# Step 2: Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Subplot 1: Line plot
axs[0, 0].plot(x, y, color='red', linestyle='-', linewidth=2, label='sin(x)')
axs[0, 0].set_title("Line Plot: sin(x)")
axs[0, 0].grid(True, linestyle='--', alpha=0.6)
axs[0, 0].legend()

# Subplot 2: Bar chart
axs[0, 1].bar(bar_categories, bar_values, color=['skyblue', 'orange', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
for i, val in enumerate(bar_values):
    axs[0, 1].text(i, val + 0.3, str(val), ha='center', fontsize=9)
axs[0, 1].grid(True, axis='y', linestyle='--', alpha=0.6)

# Subplot 3: Scatter plot
scatter = axs[1, 0].scatter(scatter_x, scatter_y, c=scatter_colors, s=100, cmap='viridis', alpha=0.8, edgecolors='black')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Subplot 4: Histogram
axs[1, 1].hist(hist_data, bins=20, color='pink', edgecolor='black', alpha=0.7)
axs[1, 1].set_title("Histogram")
axs[1, 1].grid(True, linestyle='--', alpha=0.6)

# Step 3: Overall title
fig.suptitle("2x2 Grid of Different Styled Plots", fontsize=16, fontweight='bold')

# Step 4: Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Step 5: Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame of student scores
data = {
    'Math': [88, 92, 79, 93, 85, 78, 85, 90, 86, 91],
    'Science': [84, 89, 92, 88, 76, 95, 89, 84, 91, 87],
    'English': [78, 85, 88, 90, 86, 82, 87, 89, 84, 80]
}

df = pd.DataFrame(data)

# Create a colorful boxplot
plt.figure(figsize=(8, 5))
box = df.boxplot(
    patch_artist=True,  # Fill with color
    boxprops=dict(facecolor="lightblue", color="blue"),
    whiskerprops=dict(color="blue"),
    capprops=dict(color="blue"),
    medianprops=dict(color="red", linewidth=2),
    flierprops=dict(markerfacecolor="orange", marker='o', markersize=8)
)

plt.title('Distribution of Student Exam Scores', fontsize=14, fontweight='bold')
plt.ylabel('Scores', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [30, 25, 35, 10]  

# Find index of the largest segment to explode it
largest_index = market_share.index(max(market_share))

# Explode only the largest segment
explode = [0] * len(market_share)
explode[largest_index] = 0.1  

# Create pie chart
plt.figure(figsize=(7,7))
plt.pie(
    market_share, 
    labels=companies, 
    autopct='%1.1f%%', 
    explode=explode, 
    shadow=True, 
    startangle=140,
    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99']
)
plt.title('Market Share of Companies')
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ['TechX', 'Innovatech', 'DataWorks', 'GreenSolutions']
market_share = [30, 25, 35, 10]

largest_index = market_share.index(max(market_share))
explode = [0] * len(market_share)
explode[largest_index] = 0.1

plt.figure(figsize=(7,7))
wedges, texts, autotexts = plt.pie(
    market_share,
    labels=companies,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=140,
    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],
    textprops={'color':"black", 'fontsize':12}
)

plt.title('Market Share of Companies', fontsize=16, fontweight='bold')

plt.legend(wedges, companies, title="Companies", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
x = np.arange(10)
y1 = np.random.randint(1, 10, size=10)
y2 = np.random.randint(1, 10, size=10)

plt.figure(figsize=(8,5))

# Plot two lines with labels for legend
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')

# Reverse the x-axis
plt.gca().invert_xaxis()

# Move legend outside the plot (right side) and increase font size
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14, title='Lines')

plt.title('Random Line Plot with Customized Axis and Legend')
plt.xlabel('X Axis (Reversed)')
plt.ylabel('Y Axis Values')

plt.grid(True)
plt.tight_layout()  # Adjust layout to fit legend outside
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.arange(10)
y1 = np.random.randint(1, 10, size=10)
y2 = np.random.randint(1, 10, size=10)

plt.figure(figsize=(8,5))

plt.plot(x, y1, label='Line 1', marker='o', linestyle='-', color='blue')
plt.plot(x, y2, label='Line 2', marker='s', linestyle='--', color='green')

plt.gca().invert_xaxis()

# Legend customization
plt.legend(
    loc='center left',
    bbox_to_anchor=(1, 0.5),
    fontsize=14,
    title='Lines',
    title_fontsize=16,
    shadow=True,
    fancybox=True,
    borderpad=1
)

# Axis label font customization
plt.xlabel('X Axis (Reversed)', fontsize=14, fontweight='bold', color='purple')
plt.ylabel('Y Axis Values', fontsize=14, fontweight='bold', color='purple')

plt.title('Random Line Plot with Customized Axis and Legend', fontsize=16, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
