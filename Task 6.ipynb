```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the fake jobs board
base_url = "https://realpython.github.io/fake-jobs/"

job_titles = []
companies = []
locations = []

# Loop through first 3 pages
for page in range(1, 4):
    url = f"{base_url}?page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all job cards
    job_cards = soup.find_all("div", class_="card-content")

    for job in job_cards:
        title = job.find("h2", class_="title").text.strip()
        company = job.find("h3", class_="subtitle").text.strip()
        location = job.find("p", class_="location").text.strip()

        job_titles.append(title)
        companies.append(company)
        locations.append(location)

# Create DataFrame
df = pd.DataFrame({
    "Job Title": job_titles,
    "Company": companies,
    "Location": locations
})

# Save to CSV
df.to_csv("fake_jobs.csv", index=False)

print("Data saved to fake_jobs.csv")
df.head(10)  # Show first 10 rows
```

```python
import requests
from prettytable import PrettyTable


cities = ["Mumbai", "Delhi", "London", "New York", "Tokyo"]

# Create table
table = PrettyTable()
table.field_names = ["City", "Description", "Temperature"]

for city in cities:
    try:
        # Get weather data in JSON from wttr.in
        url = f"https://wttr.in/{city}?format=j1"
        response = requests.get(url)
        data = response.json()
        
        # Extract description & temperature
        description = data['current_condition'][0]['weatherDesc'][0]['value']
        temperature = data['current_condition'][0]['temp_C'] + "°C"
        
        # Add row to table
        table.add_row([city, description, temperature])
    except Exception as e:
        table.add_row([city, "Error", "Error"])

# Display table
print(table)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://realpython.github.io/fake-jobs/"

# Storage lists
titles = []
companies = []
locations = []

# Loop through the first 3 pages
for page in range(1, 4):
    if page == 1:
        url = base_url
    else:
        url = f"{base_url}?page={page}"
    print(f"Fetching page {page}: {url}")
    
    resp = requests.get(url)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")
    
    # Find all job cards
    job_cards = soup.find_all("div", class_="card-content")
    print(f"  → Found {len(job_cards)} job listings")
    
    for card in job_cards:
        title = card.find("h2", class_="title").get_text(strip=True)
        company = card.find("h3", class_="company").get_text(strip=True)
        location = card.find("p", class_="location").get_text(strip=True)
        
        titles.append(title)
        companies.append(company)
        locations.append(location)

# Build DataFrame and save CSV
df = pd.DataFrame({
    "Job Title": titles,
    "Company": companies,
    "Location": locations
})
csv_filename = "fake_jobs_first_3_pages.csv"
df.to_csv(csv_filename, index=False)
print(f"\nSaved {len(df)} records to {csv_filename}")

# Display first 10 rows
df.head(10)
```
